```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Simulao dos dados (mesmo processo anterior)
anos = [2022, 2023, 2024]
datas = pd.date_range(start='2022-01-01', end='2024-12-31', freq='M')
regioes = ['Norte', 'Nordeste', 'Centro-Oeste', 'Sudeste', 'Sul']
empresas = ['Claro', 'Vivo', 'TIM', 'Oi', 'Sky']
perfils_clientes = ['Jovem', 'Adulto', 'Idoso']
areas_profissionais = ['Tecnologia', 'Sade', 'Educao', 'Comrcio', 'Indstria', 'Administrao']

np.random.seed(42)
chamadas_totais = np.random.randint(1500000, 3000000, size=len(datas))
chamadas_blocadas = (chamadas_totais * np.random.uniform(0.4, 0.6, size=len(datas))).astype(int)
regiao = np.random.choice(regioes, size=len(datas))
empresa = np.random.choice(empresas, size=len(datas))
perfil_cliente = np.random.choice(perfils_clientes, size=len(datas))
area_profissional = np.random.choice(areas_profissionais, size=len(datas))

df = pd.DataFrame({
    'Data': datas,
    'Chamadas_Blocadas': chamadas_blocadas,
    'Chamadas_Totais': chamadas_totais,
    'Regio': regiao,
    'Empresa_Origem': empresa,
    'Perfil_Cliente': perfil_cliente,
    'rea_Profissional': area_profissional
})

# Agrupar os dados por Regio e rea Profissional
region_profession_counts = df.groupby(['Regio', 'rea_Profissional']).agg(
    chamadas_blocadas=('Chamadas_Blocadas', 'sum')
).reset_index()

# Pivotar para barras agrupadas
pivot_data = region_profession_counts.pivot(index='rea_Profissional', columns='Regio', values='chamadas_blocadas')

# Criar grfico de barras agrupadas
pivot_data.plot(kind='bar', figsize=(12, 7))
plt.title('Chamadas Bloqueadas por rea Profissional e Regio (2022-2024)')
plt.xlabel('rea Profissional')
plt.ylabel('Nmero de Chamadas Bloqueadas')
plt.xticks(rotation=45)
plt.legend(title='Regio')
plt.tight_layout()
plt.show()
```

```python
#Previso e a categorizao de risco
# 1. Previso de chamadas bloqueadas para os prximos 6 meses
#Usando regresso linear.

# 2. Categorizao do nvel de risco por chamada (baixo, mdio, alto)
#Baseado na taxa de bloqueio por ms.


import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Simulao de dados
np.random.seed(42)
datas = pd.date_range(start='2022-01-01', end='2024-12-31', freq='M')
regioes = ['Norte', 'Nordeste', 'Centro-Oeste', 'Sudeste', 'Sul']
empresas = ['Claro', 'Vivo', 'TIM', 'Oi']
perfils_clientes = ['Jovem', 'Adulto', 'Idoso']
areas_profissionais = ['Tecnologia', 'Sade', 'Educao', 'Comrcio', 'Indstria', 'Administrao']

chamadas_totais = np.random.randint(1500000, 3000000, size=len(datas))
chamadas_blocadas = (chamadas_totais * np.random.uniform(0.4, 0.6, size=len(datas))).astype(int)
regiao = np.random.choice(regioes, size=len(datas))
empresa = np.random.choice(empresas, size=len(datas))
perfil_cliente = np.random.choice(perfils_clientes, size=len(datas))
area_profissional = np.random.choice(areas_profissionais, size=len(datas))

df = pd.DataFrame({
    'Data': datas,
    'Chamadas_Blocadas': chamadas_blocadas,
    'Chamadas_Totais': chamadas_totais,
    'Regio': regiao,
    'Empresa_Origem': empresa,
    'Perfil_Cliente': perfil_cliente,
    'rea_Profissional': area_profissional
})

# ------------------------------
# 1. PREVISO DE CHAMADAS BLOQUEADAS
# ------------------------------
df['Mes_num'] = np.arange(len(df))
X = df[['Mes_num']]
y = df['Chamadas_Blocadas']

modelo = LinearRegression()
modelo.fit(X, y)

futuro = pd.DataFrame({'Mes_num': np.arange(len(df), len(df) + 6)})
futuro['Chamadas_Blocadas_Previstas'] = modelo.predict(futuro[['Mes_num']])
futuro['Data'] = pd.date_range(start=datas[-1] + pd.DateOffset(months=1), periods=6, freq='M')

print(" Previso de Chamadas Bloqueadas para os Prximos 6 Meses:")
print(futuro[['Data', 'Chamadas_Blocadas_Previstas']])

# ------------------------------
# 2. CATEGORIZAO DE RISCO
# ------------------------------
df['Taxa_Bloqueio'] = df['Chamadas_Blocadas'] / df['Chamadas_Totais']
df['Nvel_Risco'] = pd.cut(df['Taxa_Bloqueio'], bins=[0, 0.45, 0.55, 1], labels=['Baixo', 'Mdio', 'Alto'])

print("\n Exemplo de dados com nvel de risco:")
print(df[['Data', 'Chamadas_Blocadas', 'Chamadas_Totais', 'Taxa_Bloqueio', 'Nvel_Risco']].head())

# ------------------------------
# GRFICO DAS CHAMADAS BLOQUEADAS + PREVISO
# ------------------------------
plt.figure(figsize=(14, 8))
plt.plot(df['Data'], df['Chamadas_Blocadas'], label='Chamadas Bloqueadas - Histrico', color='red')
plt.plot(futuro['Data'], futuro['Chamadas_Blocadas_Previstas'], label='Previso - Prximos 6 meses', color='orange', linestyle='--', marker='o')
plt.xlabel('Data')
plt.ylabel('Chamadas Bloqueadas')
plt.title('Previso de Chamadas Bloqueadas + Nvel de Risco')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
