```python
#Previsão e a categorização de risco
#✅ 1. Previsão de chamadas bloqueadas para os próximos 6 meses
#Usando regressão linear.

#✅ 2. Categorização do nível de risco por chamada (baixo, médio, alto)
#Baseado na taxa de bloqueio por mês.


import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Simulação de dados
np.random.seed(42)
datas = pd.date_range(start='2022-01-01', end='2024-12-31', freq='M')
regioes = ['Norte', 'Nordeste', 'Centro-Oeste', 'Sudeste', 'Sul']
empresas = ['Claro', 'Vivo', 'TIM', 'Oi']
perfils_clientes = ['Jovem', 'Adulto', 'Idoso']
areas_profissionais = ['Tecnologia', 'Saúde', 'Educação', 'Comércio', 'Indústria', 'Administração']

chamadas_totais = np.random.randint(1500000, 3000000, size=len(datas))
chamadas_blocadas = (chamadas_totais * np.random.uniform(0.4, 0.6, size=len(datas))).astype(int)
regiao = np.random.choice(regioes, size=len(datas))
empresa = np.random.choice(empresas, size=len(datas))
perfil_cliente = np.random.choice(perfils_clientes, size=len(datas))
area_profissional = np.random.choice(areas_profissionais, size=len(datas))

df = pd.DataFrame({
    'Data': datas,
    'Chamadas_Blocadas': chamadas_blocadas,
    'Chamadas_Totais': chamadas_totais,
    'Região': regiao,
    'Empresa_Origem': empresa,
    'Perfil_Cliente': perfil_cliente,
    'Área_Profissional': area_profissional
})

# ------------------------------
# 1. PREVISÃO DE CHAMADAS BLOQUEADAS
# ------------------------------
df['Mes_num'] = np.arange(len(df))
X = df[['Mes_num']]
y = df['Chamadas_Blocadas']

modelo = LinearRegression()
modelo.fit(X, y)

futuro = pd.DataFrame({'Mes_num': np.arange(len(df), len(df) + 6)})
futuro['Chamadas_Blocadas_Previstas'] = modelo.predict(futuro[['Mes_num']])
futuro['Data'] = pd.date_range(start=datas[-1] + pd.DateOffset(months=1), periods=6, freq='M')

print("🔮 Previsão de Chamadas Bloqueadas para os Próximos 6 Meses:")
print(futuro[['Data', 'Chamadas_Blocadas_Previstas']])

# ------------------------------
# 2. CATEGORIZAÇÃO DE RISCO
# ------------------------------
df['Taxa_Bloqueio'] = df['Chamadas_Blocadas'] / df['Chamadas_Totais']
df['Nível_Risco'] = pd.cut(df['Taxa_Bloqueio'], bins=[0, 0.45, 0.55, 1], labels=['Baixo', 'Médio', 'Alto'])

print("\n📊 Exemplo de dados com nível de risco:")
print(df[['Data', 'Chamadas_Blocadas', 'Chamadas_Totais', 'Taxa_Bloqueio', 'Nível_Risco']].head())

# ------------------------------
# GRÁFICO DAS CHAMADAS BLOQUEADAS + PREVISÃO
# ------------------------------
plt.figure(figsize=(14, 8))
plt.plot(df['Data'], df['Chamadas_Blocadas'], label='Chamadas Bloqueadas - Histórico', color='red')
plt.plot(futuro['Data'], futuro['Chamadas_Blocadas_Previstas'], label='Previsão - Próximos 6 meses', color='orange', linestyle='--', marker='o')
plt.xlabel('Data')
plt.ylabel('Chamadas Bloqueadas')
plt.title('Previsão de Chamadas Bloqueadas + Nível de Risco')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Esse módulo de previsão e categorização de risco é um passo importante para transformar a análise em uma ferramenta estratégica e preventiva.

```python
!pip install plotly

import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

# Simulando novamente os dados se necessário
np.random.seed(42)
datas = pd.date_range(start='2022-01-01', end='2024-12-31', freq='M')
regioes = ['Norte', 'Nordeste', 'Centro-Oeste', 'Sudeste', 'Sul']
empresas = ['Claro', 'Vivo', 'TIM', 'Oi']
perfils_clientes = ['Jovem', 'Adulto', 'Idoso']
areas_profissionais = ['Tecnologia', 'Saúde', 'Educação', 'Comércio', 'Indústria', 'Administração']

df = pd.DataFrame({
    'Data': datas,
    'Chamadas_Totais': np.random.randint(1500000, 3000000, len(datas)),
    'Região': np.random.choice(regioes, len(datas)),
    'Empresa_Origem': np.random.choice(empresas, len(datas)),
    'Perfil_Cliente': np.random.choice(perfils_clientes, len(datas)),
    'Área_Profissional': np.random.choice(areas_profissionais, len(datas))
})

df['Chamadas_Blocadas'] = (df['Chamadas_Totais'] * np.random.uniform(0.4, 0.6, len(df))).astype(int)
df['Taxa_Bloqueio'] = df['Chamadas_Blocadas'] / df['Chamadas_Totais']
df['Nível_Risco'] = pd.cut(df['Taxa_Bloqueio'], bins=[0, 0.45, 0.55, 1], labels=['Baixo', 'Médio', 'Alto'])

# ---------------------
# 1. MAPA DE CALOR POR REGIÃO E EMPRESA
# ---------------------
heatmap_data = df.groupby(['Região', 'Empresa_Origem'])['Taxa_Bloqueio'].mean().reset_index()
heatmap_pivot = heatmap_data.pivot(index='Região', columns='Empresa_Origem', values='Taxa_Bloqueio')

fig_heatmap = px.imshow(
    heatmap_pivot,
    text_auto=True,
    color_continuous_scale='Reds',
    labels=dict(color='Taxa de Bloqueio'),
    title='Mapa de Calor: Risco Médio por Região e Empresa'
)
fig_heatmap.show()

# ---------------------
# 2. ALERTAS: RISCO ALTO (> 0.55)
# ---------------------
alertas = df[df['Taxa_Bloqueio'] > 0.55][['Data', 'Região', 'Empresa_Origem', 'Taxa_Bloqueio']].sort_values(by='Taxa_Bloqueio', ascending=False)
print("🚨 ALERTAS DE RISCO ELEVADO (Taxa de Bloqueio > 55%):")
print(alertas.head(10))

# ---------------------
# 3. GRÁFICO DE BARRAS INTERATIVO: BLOQUEIOS POR EMPRESA
# ---------------------
empresa_blocos = df.groupby('Empresa_Origem')['Chamadas_Blocadas'].sum().reset_index()

fig_barras = px.bar(
    empresa_blocos,
    x='Empresa_Origem',
    y='Chamadas_Blocadas',
    title='Total de Chamadas Bloqueadas por Empresa',
    color='Chamadas_Blocadas',
    color_continuous_scale='Bluered'
)
fig_barras.show()
```
